### 0. How are businesses with different stars distributed in cities in WA?
#### Sanity checks
Before we further analyze the business data in Washington state, we need to check the data quality. After extracting datapoints with `'state'=='WA'`, we find that there're 3121 datapoints with listed cities `'North Vancouver', 'Vancover', 'Felida', 'Vancouver', 'Castle Rock', 'Vanvouver', 'Camas', 'Orchards', 'Everett', 'Portland', 'Ridgefield', 'Beaverton', 'Homan', 'VANCOUVER', 'Hazel Dell'`. Since North Vancouver belongs to Canada and Beaverton and Portland are in Oregon state, here we exclude these datapoints with these three cities. We also combine datapoints with city `'Vancover', 'Vancouver', 'Vanvouver', and 'VANCOUVER'` into `'Vancouver'`, as we believe these are typos or case negligence. After this revision, we get 3115 business datapoints in WA. Now let's visulize it!

Before this plotting, we didn't expect that the business data in WA is so incomplete. This Yelp dataset might have done some data filtering or other processing so that an overwhelming majority of businesses are located in Vancouver and it's impossible to grasp a sense of distribution as one can tell the data is not fair by one glance. Hence, we're not going to do further analysis in the aspect of businesses distribution. We'll keep on users and reviews data.

This question naturally comes up to my mind when I am thinking: What makes a customer want to write a review? Is it because it's super good that one would like to advertise it? Or is it because it's too bad that one wants to complain? Or people are willing to write reviews any time? Driven by this thought, we're going to explore the review data to see if there's a relationship.

Assume the dataset is not collected biasedly in the aspect of reviews, it seems that people tend to write a review when they think the business is good (4&5 star) or bad (1 star).

Not all the reviews takes the same time. Maybe leaving a long review will show people's willingness more strongly. Hence we'd like to further investigate the text people leave.

Although people tend to leave review for very good (4&5 stars) and very bad (1 star) business, this plot shows that people like to leave longer review when they give less stars. Do people tend to say more in a complaint rather than a compliment? (I have no idea but it's definitely interesting to me!)

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
business_data = []
for line in open("yelp_dataset/yelp_academic_dataset_business.json",'r'):
    business_data.append(json.loads(line))

cat = {}
for b_data in business_data:
    categories = b_data['categories']
    if categories is not None:
        categories = categories.split(', ')
    else:
        continue
    for c in categories:
        if c in cat:
            cat[c] = cat[c] + b_data['review_count']
        else:
            cat[c] = b_data['review_count']
```

```python
cat = dict(sorted(cat.items(), key=lambda item: item[1], reverse=True))
print(cat)
```

```python
labels = list(cat.keys())[0:15]
# labels.append('Others')
cnt = list(cat.values())[0:15]
# cnt.append(sum(list(cat.values())[20:]))
print(labels)
print(cnt)
```

```python
df = pd.DataFrame([labels, cnt]).transpose()
df.columns = ['Categories','Reviews Count']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,12))
g=sb.barplot(x="Reviews Count", y="Categories", data=df, palette=sb.color_palette())
for p in g.patches:
    g.annotate('{}'.format(int(p.get_width())), (p.get_width()-200000, p.get_y()+0.5), ha="center", size=16)
# for index, row in g.patches:
#     g.annotate(cnt[index], row, round(row.total_bill,2), color='black', ha="center")
# fig, ax = plt.subplots(figsize=(18,12))
# def make_autopct(values):
#     def my_autopct(pct):
#         total = sum(values)
#         val = int(round(pct*total/100.0))
#         return '{p:.2f}%\n({v:d})'.format(p=pct,v=val)
#     return my_autopct
# ax.pie(cnt, explode=[0.1]*15, labels=labels, autopct=make_autopct(cnt))
plt.title('Top 15 Reviewed Business')
plt.savefig('TopReviewed.png')
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory
import matplotlib.pyplot as plt

import collections
import re, string
import sys
import time

from subprocess import check_output
%matplotlib inline
import json

def init_ds(json):
    ds= {}
    keys = json.keys()
    for k in keys:
        ds[k]= []
    return ds, keys

def read_json(file):
    dataset = {}
    keys = []
    with open(file) as file_lines:
        for count, line in enumerate(file_lines):
            data = json.loads(line.strip())
            if count ==0:
                dataset, keys = init_ds(data)
            for k in keys:
                dataset[k].append(data[k])
                
        return pd.DataFrame(dataset)
```

```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
import json
import pandas as pd
import seaborn as sb
business_data = []
for line in open("yelp_dataset/yelp_academic_dataset_business.json",'r'):
    business_data.append(json.loads(line))

print(business_data[0])

# Sanity check
WA_business = [bdata for bdata in business_data if bdata['state']=='WA' and bdata['city']!='North Vancouver' and bdata['city']!='Portland' and bdata['city']!='Beaverton']
print(len(WA_business))
print(WA_business[0])

# Combine 'Vancouver'
for b in WA_business:
    if b['city']=='Vancover' or b['city']=='Vanvouver' or b['city']=='VANCOUVER':
        b['city']='Vancouver'
        
# convert to csv
out_file = open("WA_business.json","w")
json.dump(WA_business, out_file)
out_file.close()

df = pd.read_json(r'WA_business.json')
df.to_csv(r'WA_business.csv',index=None,header=True)

# plot
plt.figure(figsize=(10,6))
g=sb.countplot(data=df, x='city', hue='stars')
plt.xticks(rotation=300)
g.set_title('Business count by stars and city')
plt.savefig('Business Count in WA.png')
```
